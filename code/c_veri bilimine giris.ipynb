```python
#seaborn k√ºt√ºphanesi
#g√ºn ve saate g√∂re ƒ±sƒ± haritasƒ± (heatmap) olu≈üturma
df_a=df[df['Branch']=='A']
df_a
```

```python
hm_data=pd.crosstab(df_a['Day'],df_a['Time'])
hm_data
```

```python
#g√ºnleri sƒ±ralamak
hm_data=hm_data.reindex(['Monday', 'Tuesday', 'Wednesday','Thursday','Friday','Saturday','Sunday'])
hm_data
```

```python
sns.heatmap(hm_data, annot=True, cmap='YlGnBu')
```

```python
plt.style.use('seaborn-v0_8-whitegrid') # grafik stili
plt.figure(figsize=(18,8)) # grafiƒüin boyutlarƒ± #ebat
sns.heatmap(hm_data, annot=True, cmap='YlGnBu') # grafiƒüi √ßizdir
plt.title('G√ºn-saat i≈ülem yoƒüunluƒüu') #grafiƒüin ba≈ülƒ±ƒüƒ±
plt.xlabel('Saat', fontsize=12) #x eksen etiket
plt.ylabel('G√ºn', fontsize=12) #y eksen etiket
plt.xticks(rotation=45) # yazƒ±larƒ±n a√ßƒ±sƒ±
plt.yticks(rotation=45) #yazƒ±larƒ±n a√ßƒ±sƒ±
plt.show()# eklediƒüimiz yeni ba≈ülƒ±k ve √∂zellikleir grafikt g√∂stermek i√ßin kullanƒ±lƒ±r
```
